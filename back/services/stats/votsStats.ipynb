```python
import json
import matplotlib.pyplot as plt
import os
from datetime import datetime

# Llegir el fitxer vots.json
with open('vots.json', 'r') as file:
    vots = json.load(file)

# Crear directori si no existeix
if not os.path.exists('statImages'):
    os.makedirs('statImages')

# Comptar els vots
vots_positius = sum(1 for vot in vots if vot['vot'] == True)
vots_negatius = sum(1 for vot in vots if vot['vot'] == False)
idAct = vots[0]['idActivitat']

# Dades per al gràfic de roda
labels = 'Vots a Favor', 'Vots en Contra'
sizes = [vots_positius, vots_negatius]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  # Explode el primer slice

# Crear el gràfic de roda
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

plt.axis('equal')  # Assegurar que el gràfic és un cercle
plt.title('Resultats de les Votacions\n')

# Genera un nom únic per al fitxer basat en la data i l'hora actuals
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f'statImages/vots/{idAct}.png'

# Guarda el gràfic a la carpeta statImages
plt.savefig(filename)

plt.show()
```
